```python
import numpy as np
import pandas as pd
import os
import random
def set_seed(seed: int):
    random.seed(seed) # Python
    np.random.seed(seed)  # Numpy, é o gerador utilizado pelo sklearn
    os.environ["PYTHONHASHSEED"] = str(seed)  # sistema operativo

set_seed(25)
```

```python
import pandas as pd

#df = pd.read_csv("human_or_ai_dataset.csv")
df_treino = pd.read_csv("train.csv")
#df_teste = pd.read_csv("test.csv")
#df_validacao = pd.read_csv("validate.csv")
```

```python
variances = df_treino.var().sort_values()
print(variances.head(20))  # Mostrar as 20 colunas com menor variância
```

```python
sparsity = (df_treino == 0).sum().sum() / (df_treino.shape[0] * df_treino.shape[1])
print(f"Sparsidade do dataset: {sparsity:.4f}")
```

```python
print(df_treino.loc[:, (df_treino != 0).any(axis=0)].head())  
```

```python
word_frequencies = df_treino.sum().sort_values(ascending=False)
print(word_frequencies.head(20))  
```

```python
human_texts = df_treino[df_treino['label'] == 0].drop(columns=['label']).sum()
ai_texts = df_treino[df_treino['label'] == 1].drop(columns=['label']).sum()

diff = (ai_texts - human_texts).sort_values(ascending=False)  # Diferença de frequências

print("Palavras mais comuns na IA vs Humanos:")
print(diff.head(10))
```
